```python
import os
from sklearn.model_selection import train_test_split
import shutil
from pathlib import Path
```

```python
def split(path):
    train = 550
    val = 250
    test = 100
    total = train + val + test
    folder = ['train_HR', 'train_LR_bicubic_X2/DIV2K_train_LR_bicubic/X2']
    for f in folder:
        img = list((path/f).glob('*.png'))
        if (folder.index(f) == 0):
            train_files, val_files = train_test_split(img, test_size= val/(train + val))
            os.makedirs(path/'DIV2K_train_HR', exist_ok=True)
            os.makedirs(path/'DIV2K_train_LR_bicubic_X2', exist_ok=True)
            os.makedirs(path/'DIV2K_valid_HR', exist_ok=True)
            os.makedirs(path/'DIV2K_valid_LR_bicubic_X2', exist_ok=True)
            for file in train_files:
                print(file)
                print(file.name[0:4])
                shutil.move(str(file), str(path/'DIV2K_train_HR'/file.name))
            for file in val_files:
                shutil.move(str(file), str(path/'DIV2K_valid_HR'/file.name))
        else:
            train_list = list((path/'DIV2K_train_HR').glob('*.png'))
            train_list = [i.name[0:4] for i in train_list]
            for image in img:
                if (image.name[0:4] in train_list):
                    shutil.move(str(image), str(path/'DIV2K_train_LR_bicubic_X2'/image.name))
                else:
                    shutil.move(str(image), str(path/'DIV2K_valid_LR_bicubic_X2'/image.name))
```

```python
path = Path('./data')
```

```python
split(path)
```
